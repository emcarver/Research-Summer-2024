```python
time_3=[]
err_3=[]
flux_3=[] 

time_4=[]
err_4=[]
flux_4=[] 

time_6=[]
err_6=[]
flux_6=[] 

with open('datafile_GRS_1915+105_06-Jun-2023_97.48916GHz_0hours0min2.0sec.txt') as file:
        for line in file:
            space_one=line.find(' ')
            space_two=line.find(' ', space_one+1)
            space_three=line.find(' ', space_two+1)
            space_four=line.find(' ', space_three+1)
            space_five=line.find(' ', space_four+1)
            line_break=line.find('\n', space_five)

            MJD=line[0:space_one]
            int_flux=line[space_one+1:space_two]
            int_err=line[space_two+1:space_three]
            peak_flux=line[space_three+1:space_four]
            peak_err=line[space_four+1:space_five]
            rms_err=line[space_five+1:line_break]

            if MJD != '#MJD|integ':
                time_3.append(float(MJD))
            if peak_flux != 'err|peak':
                flux_3.append(float(peak_flux))
            if rms_err != 'imfit err|rms error':
                err_3.append(float(rms_err))
                
with open('datafile_GRS_1915+105_06-Jun-2023_144.9839GHz_0hours0min2.0sec.txt') as file:
        for line in file:
            space_one=line.find(' ')
            space_two=line.find(' ', space_one+1)
            space_three=line.find(' ', space_two+1)
            space_four=line.find(' ', space_three+1)
            space_five=line.find(' ', space_four+1)
            line_break=line.find('\n', space_five)

            MJD=line[0:space_one]
            int_flux=line[space_one+1:space_two]
            int_err=line[space_two+1:space_three]
            peak_flux=line[space_three+1:space_four]
            peak_err=line[space_four+1:space_five]
            rms_err=line[space_five+1:line_break]

            if MJD != '#MJD|integ':
                time_4.append(float(MJD))
            if peak_flux != 'err|peak':
                flux_4.append(float(peak_flux))
            if rms_err != 'imfit err|rms error':
                err_4.append(float(rms_err))
                
with open('datafile_GRS_1915+105_06-Jun-2023_232.9742GHz_0hours0min2.0sec.txt') as file:
        for line in file:
            space_one=line.find(' ')
            space_two=line.find(' ', space_one+1)
            space_three=line.find(' ', space_two+1)
            space_four=line.find(' ', space_three+1)
            space_five=line.find(' ', space_four+1)
            line_break=line.find('\n', space_five)

            MJD=line[0:space_one]
            int_flux=line[space_one+1:space_two]
            int_err=line[space_two+1:space_three]
            peak_flux=line[space_three+1:space_four]
            peak_err=line[space_four+1:space_five]
            rms_err=line[space_five+1:line_break]

            if MJD != '#MJD|integ':
                time_6.append(float(MJD))
            if peak_flux != 'err|peak':
                flux_6.append(float(peak_flux))
            if rms_err != 'imfit err|rms error':
                err_6.append(float(rms_err)) 
```

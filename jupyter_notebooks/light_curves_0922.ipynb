```python
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
from matplotlib import ticker
import datetime

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif'}
rc('font', **font)
```

```python
time_6_10=[]
flux_6_10=[]
err_6_10=[]

with open('datafile_GX399-4_10-Sep-2022_233GHz_0hours0min2.0sec.txt') as file:
        for line in file:
            space_one=line.find(' ')
            space_two=line.find(' ', space_one+1)
            space_three=line.find(' ', space_two+1)
            space_four=line.find(' ', space_three+1)
            space_five=line.find(' ', space_four+1)
            line_break=line.find('\n', space_five)

            MJD=line[0:space_one]
            int_flux=line[space_one+1:space_two]
            int_err=line[space_two+1:space_three]
            peak_flux=line[space_three+1:space_four]
            peak_err=line[space_four+1:space_five]
            rms_err=line[space_five+1:line_break]

            if MJD != '#MJD|integ':
                time_6_10.append(float(MJD))
            if peak_flux != 'err|peak':
                flux_6_10.append(float(peak_flux))
            if rms_err != 'imfit err|rms error':
                err_6_10.append(float(rms_err))
  


time_6_11=[]
flux_6_11=[]
err_6_11=[]

with open('datafile_GX399-4_11-Sep-2022_233GHz_0hours0min2.0sec.txt') as file:
        for line in file:
            space_one=line.find(' ')
            space_two=line.find(' ', space_one+1)
            space_three=line.find(' ', space_two+1)
            space_four=line.find(' ', space_three+1)
            space_five=line.find(' ', space_four+1)
            line_break=line.find('\n', space_five)

            MJD=line[0:space_one]
            int_flux=line[space_one+1:space_two]
            int_err=line[space_two+1:space_three]
            peak_flux=line[space_three+1:space_four]
            peak_err=line[space_four+1:space_five]
            rms_err=line[space_five+1:line_break]

            if MJD != '#MJD|integ':
                time_6_11.append(float(MJD))
            if peak_flux != 'err|peak':
                flux_6_11.append(float(peak_flux))
            if rms_err != 'imfit err|rms error':
                err_6_11.append(float(rms_err))
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(111)

ax.errorbar(time_6_10, flux_6_10, err_6_10, marker='.', linestyle='-', label='233GHz', color='darkorange')

locator0 = mdates.MinuteLocator(interval=5)
locator1 = mdates.MinuteLocator(interval=1)
ax.xaxis.set_major_locator(locator0)
ax.xaxis.set_minor_locator(locator1)
ax.legend(fontsize=20)#, loc=3
ax.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.tick_params(axis='x', which='both', top='on', bottom='on', direction='in', labelbottom=True)
ax.tick_params(axis='y', which='both', left='on', right='on', direction='in') 
ax.tick_params(axis='both', which='minor', length=5, width=0.8)
ax.tick_params(axis='both', which='major', labelsize=18, length=8, width=1)
ax.yaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(5))

ax.set_xlabel('Time on 2022/09/10 (HH:MM)',fontsize=22)
ax.set_ylabel('Flux Density (mJy bm$^{-1}$)',fontsize=22)

plt.savefig('band_6_20220910.pdf', bbox_inches='tight')
```
